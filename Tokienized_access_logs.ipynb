```python
# Count of unique values in 'Department'
data['Department'].value_counts()
```

```python
# Group by 'Department' and count entries
department_analysis = data.groupby('Department').size()
print(department_analysis)
```

```python
# Analyze access by hour
hourly_access = data['Hour'].value_counts().sort_index()
print(hourly_access)
```

```python
import matplotlib.pyplot as plt
```

```python
# Top 5 IPs by number of accesses
top_ips = data['ip'].value_counts().head()
print(top_ips)
```

```python
# Analyze the number of products per category
category_counts = data['Category'].value_counts()
print(category_counts)
```

```python
# Plot the distribution of products per category
category_counts.plot(kind='bar', title='Products per Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
# Analyze access by hour
hourly_access = data['Hour'].value_counts().sort_index()
print(hourly_access)
```

```python
# Plot hourly access
hourly_access.plot(kind='line', title='Hourly Access Pattern')
plt.xlabel('Hour')
plt.ylabel('Access Count')
plt.show()
```

```python
# Analyze unique IP addresses
unique_ips = data['ip'].nunique()
print(f"Unique IP addresses: {unique_ips}")
```

```python
# Top accessed departments
department_counts = data['Department'].value_counts()
print(department_counts)
```

```python
department_counts.plot(kind='pie', autopct='%1.1f%%', title='Access by Department')
plt.ylabel('')  # Hide y-label
plt.show()
```
